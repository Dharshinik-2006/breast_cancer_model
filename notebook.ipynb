```python
import pandas as pd

# Load the dataset
df = pd.read_csv("data.csv")

# Display first few rows
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Check column names
print(df.columns)

# Encode 'diagnosis' column (M → 1, B → 0)
df["diagnosis"] = LabelEncoder().fit_transform(df["diagnosis"])

# Separate features and target variable
X = df.drop(columns=["diagnosis"])  # Keep all other columns
y = df["diagnosis"]

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
if 'dt_model' in locals():
    # Make predictions
    y_pred = dt_model.predict(X_test)

    # Display first 10 predictions and actual values
    print("Predicted:", y_pred[:10])
    print("Actual:   ", y_test[:10].values)
else:
    print("Error: The model has not been trained. Run Step 4 first.")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Display classification report
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:\n", classification_rep)
```

```python
import pandas as pd

# Create a DataFrame with actual vs predicted values
results_df = pd.DataFrame({"Actual": y_test.values, "Predicted": y_pred})

# Save to CSV
results_df.to_csv("breast_cancer_predictions.csv", index=False)

# Download the file
from google.colab import files
files.download("breast_cancer_predictions.csv")
```
